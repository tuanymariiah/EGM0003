```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.tri import Triangulation
%matplotlib inline
import seaborn as sns
import math
from sympy import *
from IPython.display import Image
import warnings
warnings.filterwarnings('ignore')
init_printing(pretty_print=true)
 
#fig, ax = plt.subplots(1, 1)
```

```
r = np.linspace(1,4,15)
l = np.linspace(1,8,15)
a = np.linspace(1,4,15)
b = np.linspace(1,8,15)

rr, ll = np.meshgrid(r, l)
aa, bb = np.meshgrid(b, a)
plt.plot(ll, rr, aa, bb,marker = '.',markerfacecolor='red', color='black', linestyle = '-') ## plota o grafico

plt.xticks(range(-2, 13))
plt.yticks(range(-2, 6)) 
plt.xlabel('L') # label l
plt.ylabel('R') # label r
plt.title('Regiao C')
plt.text(9, -1, 'REGIAO C',fontsize=20, ha='left', wrap=True)

plt.gcf().set_size_inches(10, 7) # alterar tamanho
plt.savefig('regiaoC.png')
```

```
from matplotlib.path import Path

area = []
peri = []
for i in list(r):
  for j in list(l):
    A = math.pi* math.pow(i, 2) +2*i*j
    area.append(A)
    P = 2*math.pi*i + 2*j
    peri.append(P)
    #print('Area', A, 'perimetro', P)

aa, pp = np.meshgrid(area, peri)

plt.plot(area, peri,  '+',) ## plota o grafico
plt.xlabel('A')
plt.ylabel('P')
plt.title('Regiao D')
plt.text(90, 15, 'REGIAO D',fontsize=20, ha='left', wrap=True)

plt.gcf().set_size_inches(10, 7) # alterar tamanho

plt.savefig('regiaoD.png')
```

```

triang = Triangulation(area, peri)
# Use the triangulation's default TriFinder object.
trifinder = triang.get_trifinder()

# Setup plot and callbacks.
#plt.subplot(111, aspect='equal')
plt.triplot(triang, 'go-',markerfacecolor='m',color='k',lw=0.1 )
plt.gcf().set_size_inches(10, 7) # alterar tamanho
plt.text(100, 30, 'REGIAO D', fontsize=20, ha='left', wrap=True)
#plt.xlim([0,140])
plt.grid()
plt.show()
```

Cálculo do Valor Esperado (E)

Utilizando a lei fraca dos grandes números podemos somar os valores e dividir pelo *n*

```

e_peri = sum(peri)/len(peri)
e_area = sum(area)/len(area)

e_r = sum(r)/len(r)
e_l = sum(l)/len(l)
print('E(A):', e_area)
print('E(P):', e_peri)
print('E(R):', e_r)
print('E(L):', e_l)
```
